# Create helper function and manually join these together (ugh!)

## This could be programatically done but this is easier and quicker

```python
def generate_relational_data(item_name):
    items = training_set.get(item_name)
    item_list = [ [i, items[str(i)]] for i in items ]
    item_frame = pd.DataFrame(item_list)
    item_frame.columns = ['ID',item_name]
    return(item_frame)
```

```python
bathrooms = generate_relational_data('bathrooms')
bedrooms = generate_relational_data('bedrooms')
building_id = generate_relational_data('building_id')
created = generate_relational_data('created')
description = generate_relational_data('description')
display_address = generate_relational_data('display_address')
features = generate_relational_data('features')
latitude = generate_relational_data('latitude')
longitude = generate_relational_data('longitude')
listing_id = generate_relational_data('listing_id')
manager_id = generate_relational_data('manager_id')
price = generate_relational_data('price')
street_address = generate_relational_data('street_address')
interest_level = generate_relational_data('interest_level')
```
